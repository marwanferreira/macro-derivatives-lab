```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
scorecard = pd.DataFrame({
    "strategy": [
        "WTI Call Spread",
        "Silver Call Spread",
        "Treasury Straddle",
        "Treasury Butterfly",
        "Gold/Silver Ratio"
    ],
    "signal_strength": [8.0, 7.5, 6.5, 5.5, 4.5],
    "macro_clarity": [9.0, 7.5, 7.0, 7.5, 6.0],
    "execution_ease": [8.5, 8.0, 6.0, 5.5, 8.0],
    "pricing_attractiveness": [8.5, 8.0, 6.0, 5.5, 5.0],
    "reportability": [9.0, 8.0, 8.0, 8.5, 7.0],
    "overall_fit": [8.7, 7.8, 6.5, 6.0, 5.5]
})

scorecard
```

```python
score_cols = [
    "signal_strength",
    "macro_clarity",
    "execution_ease",
    "pricing_attractiveness",
    "reportability",
    "overall_fit"
]

scorecard["average_score"] = scorecard[score_cols].mean(axis=1)
scorecard = scorecard.sort_values("average_score", ascending=False).reset_index(drop=True)

scorecard
```

```python
plt.figure(figsize=(10, 5))
plt.bar(scorecard["strategy"], scorecard["average_score"])
plt.title("Strategy Scorecard: Average Score")
plt.ylabel("Average Score")
plt.xticks(rotation=20)
plt.grid(True, axis="y")
plt.show()
```

```python
scorecard.to_csv("../outputs/tables/strategy_scorecard.csv", index=False)
scorecard
```

```python
top_strategy = scorecard.iloc[0]["strategy"]
top_score = scorecard.iloc[0]["average_score"]

print(f"Top candidate strategy: {top_strategy}")
print(f"Average score: {round(top_score, 2)}")
```

## Scorecard conclusion

The purpose of this scorecard is not to create a fake precision ranking, but to summarize the evidence collected across the research notebooks. The best strategy should combine a strong current signal, a clean macro narrative, realistic execution, attractive structure pricing, and strong explainability for the final report.

At this stage, the WTI call spread appears to be the leading candidate, with the silver call spread as the strongest backup. The Treasury straddle remains a conditional idea that depends more on catalyst timing, while the gold/silver ratio currently looks structurally interesting but tactically weaker.
